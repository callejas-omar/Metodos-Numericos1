```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
F=np.array([-5, 1, 9, 25, 55, 105])
x=([0, 1, 2, 3, 4, 5])
```

```python
F
```

```python
type(F)
```

```python
plt.scatter(x,F, color="Blue", marker='x')
plt.title ("Gr√°fica de los datos")
plt.xlabel ("Eje x")
plt.ylabel ("Eje y")
plt.grid (color="black", linestyle='--')
plt.show()
```

```python
def f(x):
   return -5+7*x-2*x**2+x**3
```

```python
x1=np.linspace(0,5,100)
```

```python
plt.plot(x1,f(x1))
plt.scatter(x,F, color="orange", marker='x')
plt.xlabel ("Eje x")
plt.ylabel ("Eje y")
plt.grid (color="black", linestyle='--')
plt.show()
```

```python
dif= F[1] - F[0]
print(dif)
```

```python

delta1=np.array([])
for i in range(len(F)-1):
   dif= F[i+1] - F[i]
   print(dif)
   delta1=np.append(delta1,dif)
   print(delta1)
#Primeras Diferencias
```

```python
#Segundas Diferencias
delta2=np.array([])
for i in range(len(delta1)-1):
   dif= delta1[i+1] - delta1[i]
   print(dif)
   delta2=np.append(delta2,dif)
   print(delta2)
```

```python
#Terceras Diferencias
delta3=np.array([])
for i in range(len(delta2)-1):
   dif= delta2[i+1] - delta2[i]
   print(dif)
   delta3=np.append(delta3,dif)
   print(delta3)
```
